```python
%load_ext lab_black

import numpy as np
import pandas as pd
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
from scipy import stats
```

```python
data_2021 = pd.read_csv("../data/airbnb_listings_2021.csv")
data_2021.columns
```

```python
# Dropping irrelevant columns
datan = data_2021[
    [
        "price",
        "neighbourhood",
    ]
]
datan.dropna()
datan = pd.get_dummies(datan, columns=["neighbourhood"])
datan.drop(["neighbourhood_Midtown"], axis=1, inplace=True)
datan.astype(float)
datan
```

```python
# Dropping irrelevant columns
datang = data_2021[
    [
        "price",
        "neighbourhood_group",
    ]
]
datang.dropna()
datang = pd.get_dummies(datang, columns=["neighbourhood_group"])
datang.drop(["neighbourhood_group_Bronx"], axis=1, inplace=True)
datang.astype(float)
datang
```

```python
Y = datan["price"]
Y
```

```python
X1 = datan
datan.drop(["price"], axis=1, inplace=True)
X1c = sm.add_constant(X1)
X1c
```

```python
X2 = datang
datang.drop(["price"], axis=1, inplace=True)
X2c = sm.add_constant(X2)
X2c
```

```python
esta = sm.OLS(Y, X1c)
est2 = esta.fit()
est2.summary()
```

```python
estb = sm.OLS(Y, X2c)
est2 = estb.fit()
est2.summary()
```
