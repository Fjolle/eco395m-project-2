```python
%load_ext lab_black

import numpy as np
import pandas as pd
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
from scipy import stats
```

```python
data_2021 = pd.read_csv("../data/airbnb_listings_2021.csv")
data_2021.columns
```

```python
# Dropping irrelevant columns
data_2021 = data_2021[
    [
        "price",
        "neighbourhood",
    ]
]
data_2021.dropna()
data_2021 = pd.get_dummies(data_2021, columns=["neighbourhood"])
data_2021.drop(["neighbourhood_Midtown"], axis=1, inplace=True)
data_2021.astype(float)
data_2021
```

```python
Y = data_2021["price"]
Y
```

```python
X = data_2021
data_2021.drop(["price"], axis=1, inplace=True)
Xc = sm.add_constant(X)
Xc
```

```python
esta = sm.OLS(Y, Xc)
est1 = esta.fit()
est1.summary()
```
